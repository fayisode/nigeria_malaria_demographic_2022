```python
individual_house_hold_recode = pd.read_csv("../generate/individual_house_hold_recode.csv")
```

```python
def to_number(value):
    if not isinstance(value, (int, float)):
        return None
    return float(value)
```

```python
children_under_5 = copy.deepcopy(individual_house_hold_recode)
children_under_5['HV105$01 Age of household members'] = pd.to_numeric(children_under_5['HV105$01 Age of household members'], errors='coerce')
children_under_5 = children_under_5[children_under_5['HV105$01 Age of household members'].apply(to_number) <= 59]
children_under_5.dropna(subset=['HML32$01 Final result of malaria from blood smear test'], inplace=True)
```

```python
children.head(10)
```

```python
def extract_first(value):
    _splits = value.split(" ")
    _clean_value = [element for element in _splits if element != '']
    if len(_clean_value) != 3:
        print(_clean_value,_splits, value)
    if len(_clean_value) > 1:
        return _clean_value[1]
    print(_clean_value, _splits, value)
    return 0
```

```python
children['New ID'] = children['CASEID   Case Identification'].apply(extract_first)
```

```python
children_under_5['New ID']=children_under_5['HHID     Case Identification'].apply(extract_first)
```

```python
children['New ID'].head()
```

```python
children_under_5['New ID'].head()
```

```python
new_children = copy.deepcopy(children)
new_house_hold = copy.deepcopy(children_under_5)
```

```python
# Add an index within each group of identical 'New ID' values
new_children['index'] = new_children.groupby('New ID').cumcount()
new_house_hold['index'] = new_house_hold.groupby('New ID').cumcount()

# Merge on both 'New ID' and the added index
merged_table = pd.merge(new_children, new_house_hold[['New ID', 'HML32$01 Final result of malaria from blood smear test', 'index']],
                        on=['New ID', 'index'], how='left', suffixes=('_children', '_household'))

# Drop the added index
merged_table = merged_table.drop('index', axis=1)
```
