```python
data = copy.deepcopy(household_recode)
```

```python

# Initialize lists to store the column names
all_with_index = []
all_with_46_index = []
all_with_46_full_index = []
all_with_13_index = []

# Loop through column names
for col in data.columns:
    split_ = col.split('$')[0]

    # Check if the column has '$01' index
    if '$01' in col:
        all_with_index.append(split_)

    # Check if the column has '$46' index
    if '$46' in col:
        # If the column has '$46' index, remove its corresponding '$13' index if it exists
        all_with_13_index = [x for x in all_with_13_index if x != split_]
        all_with_46_index.append(split_)
        all_with_46_full_index.append(col)

    # Check if the column has '$13' index
    if '$13' in col:
        all_with_13_index.append(split_)
```

```python
main_cols = [x for x in data.columns if x.split('$')[0] not in all_with_index]
index_with_46 = [x for x in data.columns if x.split('$')[0] in all_with_46_index]
```

```python
print('Other existing index', [x for x in all_with_index if x not in all_with_46_index if x not in all_with_13_index])
```

```python
print('Length of all_with_index: ', len(all_with_index))
print('Length of all_with_46_index: ', len(all_with_46_index))
print('Length of all_with_13_index: ', len(all_with_13_index))
```

```python
value_vars = index_with_46[0:46]
name = value_vars[0]
added_column_names = [name]
# Melt the DataFrame, excluding the 'column' column
df_melted_1 = data.melt(value_vars=value_vars, id_vars=main_cols, var_name='row_key', value_name=name)

# Sort the melted DataFrame by 'row_key'
df_melted_1.sort_values(by='Unnamed: 0', inplace=True)
```

```python
row_call = 1;
for i in range(len(all_with_46_index) -1):
    index = 46 * row_call
    value_vars = index_with_46[index:(index * 2)]
    name = value_vars[0]
    added_column_names.append(name)
    # Melt the DataFrame, excluding the 'column' column
    df_melted = data.melt(value_vars=value_vars, id_vars='Unnamed: 0', var_name='row_key', value_name=name)
    # Sort the melted DataFrame by 'row_key'
    df_melted.sort_values(by='Unnamed: 0', inplace=True)
    df_melted_1[name] = df_melted[name]
    row_call += 1
```

```python
#removing columns with all values null
df_melted_1.dropna(axis=1, how='all', inplace=True)
```
